```python
# dropping unneccesary columns
# in pandas axis=0 represents rows (default) and axis=1 represents columns.
df_can.drop(['AREA','REG','DEV','Type','Coverage'], axis=1, inplace=True)
df_can.head(2)
```

```python
# renaming columns
df_can.rename(columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region', 'DevName':'RegionDev'} , inplace = True)
df_can.columns
```

```python
# add a new col name Total and calculate sum of immigrants by country
df_can['Total'] = df_can.sum(axis=1)
# print the first 10 rows of the column Total
print(df_can[:10]['Total'])
```

## Visualization Data using Matplotlib

```python
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# check if matplotlib is loaded
print ('Matplotlib version: ', mpl.__version__) 
```

```python
print(plt.style.available)
mpl.style.use(['ggplot']) # applying style to matplotlib, which is ggplot
```

```python
# create a variable years to store all the years columns for easier call
years = list(map(str, range(1980, 2014)))
```

```python
years[10:]
```

##### In 2010, Haiti suffered a catastrophic magnitude 7.0 earthquake. The quake caused widespread devastation and loss of life and aout three million people were affected by this natural disaster. As part of Canada's humanitarian effort, the Government of Canada stepped up its effort in accepting refugees from Haiti. We can quickly visualize this effort using a `Line` plot:

```python
# Extract data series for Haiti, exclude the Total columns and only show the years columns
haiti = df_can.loc['Haiti', years]
haiti.head(5)
```

```python
# Plot Hati data
haiti.plot()
```

```python
haiti.index = haiti.index.map(int) # change the index values of Haiti to type integer for plotting
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show() # show the updates made to the figure
```

#### We can clearly notice how number of immigrants from Haiti spiked up from 2010 as Canada stepped up its efforts to accept refugees from Haiti. Let's annotate this spike in the plot by using the `plt.text(x, y, label)` method.

```python
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

# annotate the 2010 Earthquake. 
# syntax: plt.text(x, y, label)
plt.text(2000, 6000, '2010 Earthquake')

plt.show() 
```

```python
# Extract and plot data as the same for China and India
df_CI = df_can.loc[['India', 'China'], years]
df_CI.head()
```

```python
# Plot immigrant data for China and India
# df_CI.plot(kind='line') --> this doesn't work because x axis is currently County and y axis is years, so pandas will plot individually lines for each year
# to fix this, must use transpose() method to switch/swap row and columns
df_CI = df_CI.transpose()
df_CI.head()
```

```python
# Now plot the data with 2 lines representing China and India over the years
df_CI.index = df_CI.index.map(int) # change the index values of df_CI to type integer for plotting
df_CI.plot(kind='line')

plt.title('Immigrants from China and India 1980-2013')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')
plt.show()
```

The chart above shows India and China have very similar immigration trends from 1980 - 2013.
